# .loc syntax

`df.loc[row_location, column_name]`

```python
df.loc[ [12,13,14,15] , :]
```

```python
df.loc[ 12:15, :]
```

```python
# df.loc use garda last ko index linxa for eg. df.loc[12:15, :] garda 15 index pani linxa/print hunxa
```

```python
df.loc[12:15, 'time' : 'size']
```

```python
# note df.loc use garda column name nai lekhnu parcha column ko index lekhna paudaina
```

# iloc

`df.iloc[row_index, column_index] `

```python
df.iloc[12:15, 0:3] # iloc ma column name lekhnu pardaina
```

```python
# indexing garda last ko index ko value auudaina
```

```python
# row number 8, 9, 10 and column chai last 3 columns
```

```python
df.iloc[8:11, 4:]
```

# on average who tips more smoker or non-smoker

```python
smoker = df.loc[df['smoker'] == 'Yes']
smoker['tip'].mean()
```

```python
non_smoker = df.loc[df['smoker'] == 'No']
non_smoker['tip'].mean()
```

```python
df.groupby(['smoker']) ['tip'].mean()
```

```python
# on average dinner ma badi tip painxa ki lunch ma
```

```python
df.groupby(['time']) ['tip'].mean()
```

```python
df
```

```python
# kun day on average total_bill spend hunxa
```

```python
df.groupby(['day']) ['total_bill'].mean()
```

```python
df.groupby(['sex', 'smoker']).tip.mean()
```
